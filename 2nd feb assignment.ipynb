Q1. Explain with an example each when to use for loop and a while loop.

Ans. For loop is a control flow statement that executes code repeatedly for a particular number of iteration and in this the keyword is used is 'for',
this loop is basically used when the number of iterations is already known to us.for example:

```python
n = [1,2,3,4,5,6]
for i in n:
    print(i)
```

while loop is a loop that executs a single statement or a group of statements for the given true condition, keyword used in while loop is 'while'.It is used when the number of iteration are unknown.for example:


```python
n = int(input("enter a number: "))
i = 1
while i < n:
    print(i)
    i += 1
        
```

Q2. Write a python program to print the sum and the product of the first 10 natural numbers using for loop and while loop.


Ans. for loop:

```python
sum = 0
prod = 1
for i in range(1, 11):
    sum = sum + i
    prod = prod * i
    print(sum, prod)
    
        
    
```

while loop:

```python
sum = 0
prod = 1
i = 1
while i <= 10:
    sum = sum + i
    prod = prod * i
    print(sum, prod)
    i = i+1
```

```python
def divisible(x):
    if ((x%4 == 0) and (x%5 == 0)):
        return x
l = []
for i in range(1, 101):
    a = i**3
    divisible(a)
    l.append(i)
    i = i + 1

c = list(filter(divisible, l))
print(c)
```
