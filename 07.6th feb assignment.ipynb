```python
def flat1(x):
    l=[]
    for i in x :
        # print(i)
        if type(i)==int:
            l.append(i)
        # Insde list | Tuple | set
        elif type(i)==list or type(i)==tuple or type(i)==set:
            for j in i :
                if type(j)==int:
                    # print(j,end=" | ")
                    l.append(j)
        # Insert Dict
        elif type(i)==dict:
            # For Key
            for j in i.keys():
                if type(j)==int:
                    l.append(j)
            # For Value
            for j in i.values():
                if type(j)==list or type(j)==tuple:
                    for k in j:
                        l.append(k)
                elif type(j)==int:
                    l.append(j)
    else :
        return l

print(flat1(list1))

def Product(lst):
    product = 1
    for i in flat1(list1):
        # print(product)
        product *= i
    return product

Product(list1)
```
