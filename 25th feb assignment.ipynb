Ans. Loc and iloc are two functions in Pandas that are used to slice a data set in a Pandas DataFrame. The function  loc is typically used for label indexing or can say that named indexes given by the user and can access multiple columns, while  iloc is used for integer indexing or can say that inbuilt indexes.

Ans. The window functions in Pandas are used to perform some sort of calculations across a set of rows that are interrelated with the current row. The windows functions help us to perform the aggregated calculations across various partitions or cross-sections of the data.

A question that comes to our mind after reading the windows function is what exactly is a window? Well, a window is a value (in terms of rows) that tells us the distance between our current row and the row which is to be used for calculations.

```python
import pandas as pd

#prompt the user to enter the first date
date1 = input("enter the date in the format of (yyyy-mm-dd): ")

#prompt the user to enter the second date
date2 = input("enter the date in the format of (yyyy-mm-dd): ")


#convert the dates into datetime objects
date1 = pd.to_datetime(date1)
date2 = pd.to_datetime(date2)


#difference between days, hours, minutes
time_difference = date2 - date1
days = time_difference.days

# Display the result
print(f"The difference between {date1.date()} and {date2.date()} is {days} days")

```

```python
import pandas as pd
#prompt the user to enter file path
file_path = input("enter file path: ")

#prompt the user to enter column name
column_name = input("enter column_name: ")

#read csv file
df = pd.read_csv(file_path)

#prompt the user to enter category order
category_order = input("enter the category of data (comma seperated values): ")
category_order = [cat.strip() for cat in category_order.split(",")]

# Convert the specified column to categorical datatype
df[column_name] = pd.Categorical(df[column_name], categories=category_order, ordered=True)

# Sort the data based on the categorical column
df_sorted = df.sort_values(column_name, ascending=False)
# Display the sorted data and entered data
print('File Path Entered :',file_path)
print('Column name Entered :',column_name)
print('Category Order :',category_order,'\n')
print('==================================\n')
print(df_sorted)
```

```python
import pandas as pd

try:
    # Prompt the user to enter the file path
    file_path = input("Enter the file path: ")
    
    # Read the CSV file into a DataFrame
    df = pd.read_csv(file_path)

    # Convert the date column to datetime datatype
    df['Date'] = pd.to_datetime(df['Date'])

    # Group the data by product and date, and sum the sales
    df_grouped = df.groupby(['Product', 'Date']).sum().reset_index()

    # Pivot the data to create a matrix of sales by product and date
    df_pivot = df_grouped.pivot(index='Date', columns='Product', values='Sales')

    # Print the filepath provided
    print('File path Entered : ', file_path,'\n')
    print('=====================================\n')

    print('Input CSV provided\n')
    print(df)

    print('\n=====================================\n')
    # Plot stacked barchart overtime
    df_pivot.plot(kind='bar',
                  stacked=True,
                  title='Product Sales Over time', 
                  xlabel='Date',
                  ylabel='Sales',
                  figsize=(10,10))

except Exception as e:     
    print('File path Entered : ', file_path,'\n')
    print('Exception occured and Handled : ',e)
```

```python
import pandas as pd

file_path = input("enter the file path: ")
df = pd.read_csv(file_path)


#calculating mean ,median and mode
mean = df['Test_score'].mean()
median = df['Test_score'].median()
mode = df['Test_score'].mode()[0]

#display statistical calculation in dataframe format
df_stats = pd.DataFrame({'statistics': ['mean', 'median', 'mode'],
                         'values': [mean, median, mode]})

# Print File Path Entered:
print('File Path Entered :',file_path)
print('\n==============================\n') 
print('DataFrame Provided\n')
print(df)
print('\n==============================\n')
print('Statistics Table\n')
print(df_stats)


```
