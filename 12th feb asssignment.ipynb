Ans. In Python, we catch exceptions and handle them using try and except code blocks. The try clause contains the code that can raise an exception, while the except clause contains the code lines that handle the exception.Example:

```python
try:
    with open("python.txt", "r") as f:
        f.read()
except FileNotFoundError as e:
    print("i am handling file not found error", e)
```

Ans: Custom exceptions provide you the flexibility to add attributes and methods that are not part of a standard Python exception. These can store additional information, like an application-specific error code, or provide utility methods that can be used to handle or present the exception to a user.Custom exceptions are needed to add clear, meaningful, and user-friendly information to exceptions when errors occur while your program is running.Example:

```python
class marks_ch(Exception):
    pass
```

```python
def check_marks(marks):
    if marks < 0:
        raise marks_ch("negative marks are not valid")
    elif marks > 100:
        raise marks_ch("marks are out of the target markset")
    else:
        print("this is the valid marks")
```

```python
try:
    marks = int(input("enter marks: "))
    check_marks(marks)
except marks_ch as e:
     print(e)           
```
