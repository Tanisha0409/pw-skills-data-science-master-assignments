```python
#here i am using target guided ordinal encoding
import pandas as pd

#create a data frame of above example
df = pd.DataFrame({'age': [25,42, 37, 19, 31, 28],
                  'soft drink preference': ['coke', 'pepsi', 'mountain dew', 'coke', 'pepsi', 'coke']})

df
```

```python
#calculate the median of age for each soft drink preference
median_age = df.groupby('soft drink preference')['age'].median()
median_age
```

```python
#replace each soft drink preference to median age
df['soft_drink_preference_encoded'] = df['soft drink preference'].map(median_age)
df
```

```python
df.corr()
```

```python
df.corr(method='spearman')
```
